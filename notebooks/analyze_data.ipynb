<h2>查看训练集</h2>

```python
data = pd.read_csv("../data/wtbdata_245days.csv")
data.shape
```

```python
data.info()
```

```python
for i in range(data["TurbID"].min(), data["TurbID"].max() + 1):
    print(data[data["TurbID"] == i].shape)
```

```python
data[data["TurbID"] == 1].head(20)
```

```python
print(data[data["TurbID"] == 1]["Day"].min())
print(data[data["TurbID"] == 1]["Day"].max())
```

```python
data[data["Tmstamp"] == "00:00"]
```

```python
one_data = data[data["TurbID"] == 1]
```

```python
plt.figure(1)
plt.plot(one_data[one_data["Day"] == 1]["Patv"])
plt.show()
```

```python
plt.figure(1)
plt.plot(one_data.groupby("Tmstamp").mean()["Patv"])
plt.xticks(rotation=90)
plt.show()
```
